```python
import pandas as pd
import requests
from config import geoapify_key
```

```python
co_city_data  = pd.read_csv("Resources/CO_SingFam_TopTier_BotTier_Combined.csv")

co_city_data
```

```python
base_url = 'https://api.geoapify.com/v2/places?'
categories = 'childcare'
radius = 10000
params = {
    'categories':categories,
    'apiKey': geoapify_key,
    'limit':500
}

childcare = []

for index, row in co_city_data.iterrows():
    lat = row['Latitude']
    lon = row['Longitude']
    
    params['bias'] = f'proximity:{lon},{lat}'
    params['filter'] = f'circle:{lon},{lat},{radius}'
    
    data = requests.get(base_url, params).json()
    
    childcare.append({'City':row['City'], 'Number of Childcare Facilities':len(data['features'])})

    print(f"{row['City']}:{len(data['features'])}")
```

```python
childcare_df = pd.DataFrame(childcare)

childcare_df
```

```python
merged_df = pd.merge(childcare_df,co_city_data, on = "City", how = 'outer')

merged_df= merged_df.drop(columns = 'Unnamed: 0')
merged_df
```
